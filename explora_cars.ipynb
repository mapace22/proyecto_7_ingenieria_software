```python
# Limpieza de nombres de columna
data.columns = data.columns.str.strip().str.lower().str.replace(" ", "_")
data
```

```python
import pandas as pd

# Suponiendo que tu DataFrame se llama 'data'

columnas_object = data.select_dtypes(include=['object']).columns

for col in columnas_object:
    try:
        data[col] = data[col].str.lower()
    except AttributeError:
        print(f"Error al convertir a minúsculas la columna '{col}'. Puede que contenga valores no convertibles a string.")

# Para verificar los cambios (opcional)
for col in columnas_object:
    print(f"Valores únicos en '{col}' después de convertir a minúsculas:")
    print(data[col].unique()[:20]) # Mostrar los primeros 20 valores únicos

print(data.head()) # Para ver las primeras filas del DataFrame con los cambios
```

```python
def tienen_decimales(serie):
    """Verifica si una Serie numérica (float) tiene valores con decimales distintos de cero."""
    return (serie % 1 != 0).any()

columnas_float = data.select_dtypes(include=['float64']).columns

print("Columnas de tipo float64 y si contienen decimales distintos de cero:")
for col in columnas_float:
    if col in data.columns:
        tiene = tienen_decimales(data[col].dropna()) # Excluir NaN para evitar errores
        print(f"- {col}: {tiene}")
```

```python
import pandas as pd

# Convertir 'price' a float64
data['price'] = data['price'].astype('float64')

# Convertir 'model_year' a Int64 (entero nullable)
data['model_year'] = data['model_year'].astype('Int64')

# Convertir 'cylinders' a Int64 (entero nullable)
data['cylinders'] = data['cylinders'].astype('Int64')

# Convertir 'odometer' a Int64 (entero nullable)
data['odometer'] = data['odometer'].astype('Int64')

# Convertir 'is_4wd' a boolean (manejando NaN como False - reconsiderar si la falta de info no implica False)
data['is_4wd'] = data['is_4wd'].fillna(False)
data['is_4wd'] = data['is_4wd'].astype('bool')

# Convertir 'date_posted' a datetime64[ns]
data['date_posted'] = pd.to_datetime(data['date_posted'])

# 'days_listed' ya está en int64, no necesita conversión

# 'paint_color'está como 'object' pero se rellena con 'No Disponible' para la lógica backend
data['paint_color'] = data['paint_color'].fillna('No Disponible')

# Verificar
print(data.info())

# Filas
print(data.head(10))
```
